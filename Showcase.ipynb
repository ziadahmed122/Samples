```python
import pandas as pd 
import biotite.sequence as biotite
import Bio as Seq
from itertools import product
from Bio.Data import IUPACData
import biotite.sequence as seq
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
from pandas import DataFrame
#showing the table of translation
table = seq.CodonTable.default_table()
print('Notice that the translation will be created using this table')
print(table)
#-------------------------------------------------------------------------------
print("-------------------------------------------------------------------------")
#modifying the program to have a string that is divisible by 3
pre=input('Please input the DNA/RNA sequance of interest:')#the strand pre-treatment
pre=''.join(pre.split())#removing the spaces
while True: 
  if len(pre) % 3 != 0:
    pre=''.join([pre[i] for i in range(len(pre)) if i != len(pre)-1]) 
  else:
    break
print("-------------------------------------------------------------------------")
print("input after treatment",pre)
#-------------------------------------------------------------------------------
print("-------------------------------------------------------------------------")
#The CG content and denature temp
seq_dic_repeat = {i:pre.count(i) for i in pre}
seq_list_repeat = sorted(seq_dic_repeat.items())
x, y = zip(*seq_list_repeat)
plt.bar(x, y)
plt.title("The number of each base")
plt.show()
print("The CG contet of the sequance is",int(((seq_dic_repeat["C"]+seq_dic_repeat["G"])/len(pre))*100),"%")
print("The molecular weight of the sequance in Deltons is ",len(pre)*650)
#--------------------------------------------------------------------------------------------
Base_list=['TA','TG',"CA",'CG','AG','CT','AA','TT','AT','GA','TC','CC','GG','AC','GT','CG']
Temps=[0.12,.78,.78,1.44,1.29,1.29,1.04,1.04,1.27,1.66,1.66,1.97,1.97,2.04,2.04,2.7]
Energy=[0]
for i in Base_list:
    number_bonds=pre.find(i)
    Total= Energy[-1]+ number_bonds * Temps[Base_list.index(i)]
    Energy.append(Total)
print('Delta G needed to break the Sequance is',Energy[-1])   
#-------------------------------------------------------------------------------
#The normal strand 
print("-------------------------------------------------------------------------")
sequance = (seq.NucleotideSequence(pre))
protien,ORF=sequance.translate()
DF_ORF=[]
print("For strand 1:")
for i in range(len(protien)):
  DF_ORF.append([ORF[i][0],ORF[i][1],str(protien[i])])
Final_protien=sequance.translate(complete=True)
DF_ORF_table=pd.DataFrame(data=DF_ORF)
DF_ORF_table.columns = ['ORF beginning', 'ORF end','Protien translation']
display(DF_ORF_table)
print("The translation for the full strand",Final_protien)
#-------------------------------------------------------------------------------
print("-------------------------------------------------------------------------")
#the reverse strand 
sequance_rev = sequance.reverse().complement()
protien_rev,ORF_rev=sequance_rev.translate()
print("For strand -1:")
DF_ORF_rev=[]
for i in range(len(ORF_rev)):
  DF_ORF_rev.append([ORF_rev[i][0],ORF_rev[i][1],str(protien_rev[i])])
DF_ORF_rev_table=pd.DataFrame(data=DF_ORF_rev)
DF_ORF_rev_table.columns = ['ORF beginning', 'ORF end','Protien translation']
display(DF_ORF_rev_table)
Final_protien_rev=sequance_rev.translate(complete=True)
print("The translation for the full strand in reverse",Final_protien_rev)
#---------------------------------------------------------------------------------
print("-------------------------------------------------------------------------")
print('Dataframe for the Motifs hits')
homer_table=pd.read_csv('Homer')
user_input= pre
user_input=''.join(user_input.split())
main_seq = biotite.NucleotideSequence(user_input)
positions=[]
def extend_ambiguous_dna(seq):
    delimiter = IUPACData.ambiguous_dna_values
    try:
        return [ list(map("".join, product(*map(delimiter.get, seq)))) ]
    except:
        pass
for i in homer_table.values[1:]:
    delimited_data=extend_ambiguous_dna(i[1])
    normal=i[1]
    if delimited_data != []:
        if user_input.find(i[1]) != -1:
            positions.append([i[0],i[1],user_input.find(i[1])])
    else:
        for x in delimited_data:
            sub_seq = biotite.NucleotideSequence(x)
            try:
                if user_input.find(x) != -1:
                    positions.append([i[0],x,user_input.find(x)])
            except:
                pass
Motif_DF = DataFrame (positions,columns=['Name','sequance','position'])
display(Motif_DF)
#---------------------------------------------------------------------------------
print("-------------------------------------------------------------------------")
print('Map for the motifs hits')
import matplotlib.pyplot as plt
from biotite.sequence import Annotation, Feature, Location
import biotite.sequence.graphics as graphics
import biotite.sequence.graphics as graphics
strand = Location.Strand.FORWARD
feat_list=[]
for i in positions:
    tilte= i[0]
    loc_start=i[2]
    loc_end= int(i[2])+8
    type_feat='regulatory_class'
    class_feat= 'regulatory_class'
    class_feat_2 = 'regulatory_class'
    feat= Feature(class_feat
                  , [Location(loc_start, loc_end, strand)],
                {class_feat_2 : type_feat,
                 "note"             : tilte})
    feat_list.append(feat)
annotation = Annotation(feat_list)
#-------------------------------------------------------------------
fig = plt.figure(figsize=(20, 1))
ax = fig.add_subplot(111)
graphics.plot_feature_map(
    ax, annotation, multi_line=False, loc_range=(1, len(pre)),show_line_position=True)
fig.tight_layout()
plt.show()
    
for i in DF_ORF:
    tilte= ''
    loc_start=i[0]
    loc_end= i[1]
    type_feat='gene'
    class_feat= 'gene'
    class_feat_2 = 'gene'
    feat= Feature(class_feat
                  , [Location(loc_start, loc_end, strand)],
                {class_feat_2 : type_feat,
                 "note"             : tilte})
    feat_list.append(feat)
    
annotation = Annotation(feat_list)
#-------------------------------------------------------------------
print('Map for the ORFs')
fig = plt.figure(figsize=(20, 1))
ax = fig.add_subplot(111)
graphics.plot_feature_map(
    ax, annotation, multi_line=False, loc_range=(1, len(pre)),show_line_position=True)
fig.tight_layout()
plt.show()
```
